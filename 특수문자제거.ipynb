```python
import numpy as np
import pandas as pd
import re
import matplotlib.pyplot as plt
```

```python
import re
# 이모티콘 제거
def remove_emoji(sentence): 
  only_BMP_pattern = re.compile("["
        u"\U00010000-\U0010FFFF"  #BMP characters 이외
         u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"
                           "]+", flags=re.UNICODE)
  return only_BMP_pattern.sub(r'', sentence)
```

```python
#이모티콘 제거, 한글 이랑 알파벳아니면 공백처리
dataset=[]
for i in range(len(contentList)):
  dataset.append(re.sub('[^가-힣a-zA-Z\s]','',str(remove_emoji(contentList[i]))))
```

```python
dataset
```

```python
result_df = pd.DataFrame(dataset,columns=['content'])
```

```python
# 슬라이싱 결과 확인
for s in result_df["content"]:
    print(s)
    print("")
```
