```python
# 파일 삽입
df = pd.read_csv('specup_jobtalk.csv', encoding='cp949')
df.head() # 처음 5개만 형태 확인하기
```

```python
# 게시글 뒤에 공통적으로 붙는 공지, 광고 문장 슬라이싱
def slicing(content):
    try:
        result=""
        temp = content[:content.find("특정 문자 뒤의 내용 삭제")] # \u200b = 이모티콘 
        for letter in temp:
            if letter!="\n": # \n 버림
                result+=letter
        return result
    except:
        return 0
```

```python
for i in range(len(df)):
    df["내용"][i]=slicing(df["내용"][i])
```

```python
# 댓글에 공통적으로 들어간 필요없는 문장 제거

for i in range(len(df)):
    s = df['댓글'][i]
    s = s[:-64]
    df['댓글'][i] = s
```

```python
# 댓글 슬라이싱 후 확인
for s in df["댓글"]:
    print(s)
    print("")
```

```python
df.to_csv('specup_jobtalk_slicing.csv', encoding="utf-8-sig")
```
